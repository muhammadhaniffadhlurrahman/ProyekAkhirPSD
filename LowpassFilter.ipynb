**FIR JENDELA HAMMING**

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter lowpass FIR
def design_lowpass_fir(numtaps, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    taps = signal.firwin(numtaps, norm_cutoff, window='hamming', pass_zero=True)
    return taps

# Baca file WAV
input_wav_path = 'Virgoun feat. Audy - Selamat (Selamat Tinggal) (Official Lyric Video) _00_03_15-00_03_25.wav'
output_wav_path = 'output_filtered_FIRHamming.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
numtaps = 101  # Panjang filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
taps = design_lowpass_fir(numtaps, cutoff_freq, fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

**IIR JENDELA BESSEL**

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Baca file WAV
input_wav_path = 'Virgoun feat. Audy - Selamat (Selamat Tinggal) (Official Lyric Video) _00_03_15-00_03_25.wav'
output_wav_path = 'output_filteredIIRBessel.wav'
fs, data = wavfile.read(input_wav_path)

# Desain filter IIR lowpass Bessel
order = 4  # Orde filter Bessel
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
nyquist = 0.5 * fs  # Frekuensi Nyquist
normalized_cutoff = cutoff_freq / nyquist

# Dapatkan koefisien filter Bessel
b, a = signal.bessel(order, normalized_cutoff, btype='low', analog=False)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Bessel')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Impulse response
impulse_response = signal.lfilter(b, a, [1] + [0]*((len(b)-1)))
plt.subplot(4, 1, 3)
plt.plot(impulse_response)
plt.title('Impulse Response')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

**IIR JENDELA BUTTERWORTH**

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Baca file WAV
input_wav_path = 'Virgoun feat. Audy - Selamat (Selamat Tinggal) (Official Lyric Video) _00_03_15-00_03_25.wav'
output_wav_path = 'output_filteredIIRButterworth.wav'
fs, data = wavfile.read(input_wav_path)

# Desain filter IIR lowpass dengan Butterworth
order = 6  # Orde filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
b, a = signal.butter(order, cutoff_freq, 'low', fs=fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Impulse response
impulse_response = signal.impulse((b, a))
plt.subplot(4, 1, 3)
plt.plot(impulse_response[1])
plt.title('Impulse Response')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```
